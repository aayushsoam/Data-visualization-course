```python
import matplotlib.pyplot as plt
```

Bar chart simple code

```python
# bar chart 

# Subjects and their marks
subjects = ['Math', 'Science', 'English', 'History', 'Computer']
marks = [85, 90, 75, 80, 95]

# Different colors for each bar
colors = ['red', 'blue', 'green', 'orange', 'purple']

# Create bar chart
plt.bar(subjects, marks, color=colors)

# Add labels and title
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks in 5 Subjects')

# Show the chart
plt.show()
```

LINE CHART SIMPLE CODE

```python

# X और Y डेटा
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Line chart बनाना
plt.plot(x, y, color='blue', label='Line Chart', marker='o')

# Scatter chart उसी ग्राफ पर जोड़ना

# Labels और Title
plt.xlabel('X अक्ष')
plt.ylabel('Y अक्ष')
plt.title('Line Chart ')

# Legend दिखाना
plt.legend()

# ग्राफ दिखाना
plt.show()
```

SCATTER CHART SIMPLE CODE 

```python
# X और Y डेटा
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Line chart बनाना
plt.scatter(x, y, color='blue', label='Line Chart', marker='o')

# Scatter chart उसी ग्राफ पर जोड़ना

# Labels और Title
plt.xlabel('X अक्ष')
plt.ylabel('Y अक्ष')
plt.title('scatter Chart ')

# Legend दिखाना
plt.legend()

# ग्राफ दिखाना
plt.show()
```

```python
# Creating and showing a pie chart for top 10 world companies by revenue (approx. USD billions)


companies = ["Walmart", "Amazon", "State Grid", "Saudi Aramco", "Sinopec", 
             "China National Petroleum", "Apple", "UnitedHealth", "Berkshire Hathaway", "CVS Health"]
revenues = [648, 575, 546, 495, 430, 422, 383, 372, 364, 358]  # in USD billions (approx.)

plt.figure(figsize=(8,8))
plt.pie(revenues, labels=companies, autopct='%1.1f%%', startangle=90)
plt.title("Top 10 World Companies by Revenue (approx., USD billions)")
plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.

# Save image for user to download

plt.show()


```
