```python
n=5
for i in range (n):
    for j in range (i,n):
        print("*",end=' ')
    print()     
    
```

```python
n=int(input("enter no:"))
for i in range(1,n):
    # for space
    print(" "* (n-i),end="")
    for j in range(1,2*i):
        print(j,end="")
    print()
```
